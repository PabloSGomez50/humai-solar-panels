## Trabajar las columnas extras
 - Input: df
 - Output: df_active

Para trabajar con un dataset mas reducido es preferible utilizar el consumo de un solo cliente.

Eliminar las columnas de Generacion, codigo postal, Categoria y ultima columna ya que no tiene ningun dato significativo.

```python
# df_active = df.query('Customer <= 10')
# Seleccionar el customer 1 y eliminar las columnas sobrantes
n = 1
df_active = df[df['Customer'] == n]
capacity = df_active['Generator Capacity'].unique()[0]
df_active = df_active.drop(columns=['Postcode', 'Row Quality', 'Generator Capacity', 'Customer'])

df_active = df_active[df_active['Consumption Category'] != 'GG']
df_active = df_active.rename(columns={'Consumption Category': 'cat'})
df_active
```

## Melt y organizacion

- Output: df_melt

GC = General Consumption for electricity supplied all the time
(primary tariff, either inclining block or time of use rates),
excluding solar generation and controlled load supply

CL = Controlled Load Consumption (Off peak 1 or 2 tariffs)

GG = Gross Generation for electricity generated by the solar
system with a gross metering configuration, measured separately
to household loads

```python
df_melt = df_active.melt(id_vars=['cat', 'date'], var_name='time', value_name='consumo')
df_melt.head(3)
```

## Ordernar columnas

- Input: df_melt
- Output: df_order

```python
def cl_gc_column(df_mod):
  df_temp = df_mod[df_mod['cat'] == 'CL'].copy()
  serie_gc = df_mod[df_mod['cat'] == 'GC']['consumo'].copy()

  df_temp.rename(columns={'consumo': 'CL'}, inplace=True)
  df_temp.drop(columns='cat', inplace=True)
  df_temp.reset_index(drop=True, inplace=True)

  df_temp['GC'] = serie_gc.reset_index(drop=True)

  return df_temp.copy()
```

```python
def format_date(df_mod):

  df_mod['datetime'] = pd.to_datetime(df_mod['date'] + df_mod['time'], format='%d/%m/%Y%H:%M')
  df_mod.drop(columns=['date', 'time'], inplace=True)

  return df_mod.copy()
```

```python
# Creando las columnas GC y CL
df2 = cl_gc_column(df_melt)

# Formatear la hora
df_order = format_date(df2)

# Ordenar las columnas y los indices
df_order = df_order.loc[:,::-1]
df_order.sort_values('datetime', inplace=True)
df_order.reset_index(drop=True, inplace=True)

df_order
```
